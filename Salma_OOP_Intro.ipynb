```
# Calculating TAX in function of the quality
def calculate_tax(quality :str, price :int)->int: 
  total=0
  if quality == "good":
    total+=price*1.1
  elif quality == "medium":
    total+=price*1.07
  elif quality == "low":
    total+=price*1.05
  return total
  
#Calculate TAX of each pen
for pen in list_pens:
  quality = pen.get_quality()
  price = pen.get_price()
  tax = calculate_tax(quality, price)
  print(f"Pen {pen.get_brand()} pays TAX at", f"{'.'*20} rate {tax:.2f}%")  
```
