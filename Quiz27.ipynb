```python
def abundant(a, b):
  sum_a=0
  for num in range(a, b+1):
    sum_d=0
    for i in range(2, num):
      if num%i==0:
        sum_d+=i
    if sum_d>num:
      sum_a+=1

    
  return sum_a

#testing
print(abundant(1, 40))
```
