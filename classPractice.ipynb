```
class Person():
  def __init__(self, name, age):
        self.name :str = name
        self.age : int = age

  def display_name(self):
          return self.name

  def display_age(self):
          return self.age

class Student(Person):
  def __init__(self, name, age, grade):
    Person.__init__(self, name, age)
    self.grade :int = grade

  def display_grade(self):
    return self.grade

  def displayStudent(self):
    print(f"{self.display_name()} is {self.display_age()} years old and they are in grade {self.display_grade()}")

student1 = Student("Bob",26,10)
student1.displayStudent()
```

**Description:** The program below has 2 classes: BankAccount() and MinimumBalanceAccount().BankAccount() has 1 attribute "balance" and 2 methods withdraw() and deposit(). MinimumBalanceAccount() has 1 attribute "minimum_balance" and 1 method withdraw(). This program allows the user to withdraw and deposit money into/from their bank account while maintaining a minimum balance within the account. 

```
class BankAccount:
    def __init__(self):
        self.balance = 0

    def withdraw(self, amount):
        self.balance -= amount
        return self.balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

class MinimumBalanceAccount(BankAccount):
    def __init__(self, minimum_balance):
        BankAccount.__init__(self)
        self.minimum_balance = minimum_balance

    def withdraw(self, amount):
        if self.balance - amount < self.minimum_balance:
            print('Sorry, minimum balance must be maintained.')
        else:
            BankAccount.withdraw(self, amount)
```
